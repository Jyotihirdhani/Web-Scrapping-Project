```python
import pandas as pd
import requests
import re
import time
from google.colab import drive  # ✅ FIX ADDED

# -----------------------------
# MOUNT DRIVE
# -----------------------------
drive.mount('/content/drive')

# -----------------------------
# FILE PATHS
# -----------------------------
input_file = "/content/drive/MyDrive/Web Scrapping/website_test.xlsx"
output_file = "/content/drive/MyDrive/Web Scrapping/Websites_With_Reach.xlsx"

# -----------------------------
# LOAD FILE
# -----------------------------
df = pd.read_excel(input_file)

column_name = df.columns[0]
df["Monthly Visits"] = None

# -----------------------------
# SESSION SETUP
# -----------------------------
session = requests.Session()

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}

# -----------------------------
# MAIN LOOP
# -----------------------------
for index, row in df.iterrows():
    site = str(row[column_name]).strip()

    try:
        original_url = f"https://hypestat.com/info/{site}"
        proxy_url = f"https://api.allorigins.win/raw?url={original_url}"

        response = session.get(proxy_url, headers=headers, timeout=20)

        if response.status_code == 200:

            html = response.text

            match = re.search(r"Monthly\s+Visits.*?([\d,]+)", html, re.IGNORECASE)

            if match:
                visits = match.group(1)
                df.at[index, "Monthly Visits"] = visits
                print(f"[SUCCESS] {site} → {visits}")
            else:
                df.at[index, "Monthly Visits"] = "Not Found"
                print(f"[NOT FOUND] {site}")

        else:
            df.at[index, "Monthly Visits"] = f"Status {response.status_code}"
            print(f"[Error] {site} (Status Code {response.status_code})")

        time.sleep(3)

    except Exception as e:
        df.at[index, "Monthly Visits"] = "Error"
        print(f"[Failed] {site} ({e})")

# -----------------------------
# SAVE OUTPUT
# -----------------------------
df.to_excel(output_file, index=False)

print("\n✅ All done! Output saved successfully in Drive.")
```
